## Styles & contexts
Format all your notebook's graphs in one line !

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(20,10))
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
sinplot()
```

### Styles
Change your whole notebook style in one line !

```python
sns.set()
```

```python
sinplot()
```

```python
sns.set_style("dark")
sinplot()
```

```python
sns.set_style("darkgrid")
sinplot()
```

```python
sns.set_style("whitegrid")
sinplot()
```

```python
sns.set_style("white")
sinplot()
```

```python
sns.set_style("ticks")
sinplot()
```

### Contexts
Get presentation styled graphs instantly !

```python
sns.set_context("paper")
sinplot()
```

```python
sns.set_context("notebook")
sinplot()
```

```python
sns.set_context("talk")
sinplot()
```

```python
sns.set_context("poster")
sinplot()
```

## Data exploration

### univariate

```python

import seaborn as sns
sns.set(style="ticks")

# Load the example tips dataset
tips = sns.load_dataset("tips")

# Draw a nested boxplot to show bills by day and sex
sns.boxplot(x="day", y="total_bill", hue="sex", data=tips, palette="PRGn")
sns.despine(offset=10, trim=True)
```

### bivariate

```python
import seaborn as sns
sns.set(style="ticks")

df = sns.load_dataset("iris")
sns.pairplot(df, hue="species")
```
